```python
from datetime import datetime
import matplotlib.pyplot as plt
from meteostat import Point, Daily, Stations
df1 = df.copy()
```

```python
def weather(row):
    try:
        date = datetime.fromisoformat(row['transaction_date'])
        location = Point(lat=row['pos_adress_lat'], lon=row['pos_adress_lon'])
        data = Daily(location, date, date)
        data = data.fetch().loc[:, ['tavg', 'tmin', 'tmax']]
        return data.iloc[0, 0], data.iloc[0, 1], data.iloc[0, 2]
    except:
        return None, None, None
```

```python
tqdm.pandas()
df1[['tavg', 
     'tmin', 
     'tmax']] = df1.progress_apply(weather, axis=1, result_type="expand")
with open('df1_weather.pkl', 'wb') as f:
    pickle.dump(df1, f)
```

```python
df = df1
df['ismissing'] = df[['tmin','tmax']].isna().all(axis=1)
tmin_tmax_none = list(df.loc[df['ismissing'] == True].index)
df = df.drop(tmin_tmax_none, axis=0)
```

```python
df['tavg'] = df['tavg'].fillna(-1000)
df['tmin'] = df['tmin'].fillna(-1000)
df['tmax'] = df['tmax'].fillna(-1000)
```

```python
def clean_temp(row):
    tavg = row['tavg']
    tmin = row['tmin']
    tmax = row['tmax']
    if tmin != -1000 and tmax != -1000 and tavg == -1000:
        tavg = (tmin + tmax) / 2
    return tavg
```

```python
tqdm.pandas()
df['tavg'] = df.progress_apply(clean_temp, axis=1)

def clean_temp(row):
    tavg = row['tavg']
    tmin = row['tmin']
    tmax = row['tmax']
    if tmin != -1000 and tavg != -1000 and tmax == -1000:
        tmax = 2 * tavg - tmin
    return tmax

tqdm.pandas()
df['tmax'] = df.progress_apply(clean_temp, axis=1)


def clean_temp(row):
    tavg = row['tavg']
    tmin = row['tmin']
    tmax = row['tmax']
    if tmin == -1000 and tavg != -1000 and tmax != -1000:
        tmin = 2 * tavg - tmax
    return tmin

tqdm.pandas()
df['tmin'] = df.progress_apply(clean_temp, axis=1)
```
